```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv("evaluation_metrics_ResNet50.csv")
epochs = np.arange(0, 100, 1)

# Create figure and axis
fig, ax1 = plt.subplots()

# Plot loss
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss', color='tab:red')
ax1.plot(epochs, df["avg_loss"][:100], color='tab:red', label='Loss')
ax1.tick_params(axis='y', labelcolor='tab:red')

# second y-axis for r^2
ax2 = ax1.twinx()
ax2.set_ylabel('r2', color='tab:blue')
ax2.plot(epochs, df["r2"][:100], color='tab:blue', label='NCC')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title and layout
plt.title('Loss and r2 over Epochs')
fig.tight_layout()
plt.show()
```
