```python
import torch
import pandas as pd
import os
```

```python
ids = pd.read_csv("ids.csv", header=None, names=['f.eid'])

all_labels = pd.read_csv("all_labels_CMR_pretrain.csv")
print(all_labels.shape)

all_labels.head()
```

```python
ids_to_drop = [1357232, 1606024, 1756573, 1693054, 1903487, 2132506, 2344548, 2563947, 2706368, 3260781,
               3303947, 3728113, 3750605, 4113797, 4313384, 4476590, 4553691, 4575995, 4623638, 4859492,
               5405990, 5516382, 5627107]
```

```python
labels_for_cmr = all_labels.loc[all_labels['f.eid'].isin(ids['f.eid'])]
print(labels_for_cmr.shape)

labels_for_cmr = labels_for_cmr[~labels_for_cmr['f.eid'].isin(ids_to_drop)].reset_index()
print(labels_for_cmr.shape)

labels_for_cmr = labels_for_cmr.drop(columns=["Unnamed: 0", 'index'])

labels_for_cmr.head()
```

```python
pd.set_option('display.max_rows', None)
labels_for_cmr.isna().sum()[labels_for_cmr.isna().sum() > 0]
```

```python
# cols_to_drop = [f'f.241{i}.2.0' for i in range(10, 24)]
cols_to_drop = [f'f.241{i}.2.0' for i in range(18, 24)]

labels_for_cmr = labels_for_cmr.drop(columns=cols_to_drop)

labels_for_cmr.isna().sum()[labels_for_cmr.isna().sum() > 0]
```

```python
labels_for_cmr = labels_for_cmr.reset_index()

labels_for_cmr.head()
```

```python
# get the training, val, test data splits

train_ids = pd.read_csv("ids_cmr_pretrain_train.csv", header=None, names=['f.eid'])
print("train_ids shape:", train_ids.shape)

val_ids = pd.read_csv("ids_cmr_pretrain_val.csv", header=None, names=['f.eid'])
print("val_ids shape:", val_ids.shape)

test_ids = pd.read_csv("ids_cmr_pretrain_test.csv", header=None, names=['f.eid'])
print("test_ids shape:", test_ids.shape)


train_labels_for_cmr = labels_for_cmr.loc[labels_for_cmr['f.eid'].isin(train_ids['f.eid'])]
train_labels_for_cmr = train_ids.merge(train_labels_for_cmr, on='f.eid', how='left')

print("train_labels ids shape:", train_ids.shape)

val_labels_for_cmr = labels_for_cmr.loc[labels_for_cmr['f.eid'].isin(val_ids['f.eid'])]
val_labels_for_cmr = val_ids.merge(val_labels_for_cmr, on='f.eid', how='left')

test_labels_for_cmr = labels_for_cmr.loc[labels_for_cmr['f.eid'].isin(test_ids['f.eid'])]
test_labels_for_cmr = test_ids.merge(test_labels_for_cmr, on='f.eid', how='left')
```

```python
test_labels_for_cmr.head()
```

```python
# cmr_tensor[CL_idx]
output_dir = "/projects/prjs1252/CL_data"
full_tensor = torch.load(os.path.join(output_dir, "cmr_tensors_ts_all.pt"))
test_tensor = full_tensor[test_labels_for_cmr["index"]]
test_tensor = test_tensor.unsqueeze(1)
test_tensor.shape
```

```python
val_tensor = full_tensor[val_labels_for_cmr["index"]]
val_tensor = val_tensor.unsqueeze(1)

val_tensor.shape
```

```python
train_tensor = full_tensor[train_labels_for_cmr["index"]]
train_tensor = train_tensor.unsqueeze(1)

train_tensor.shape
```

```python
torch.save(train_tensor, os.path.join(output_dir, f"cmr_tensors_ts_all_train.pt"))
torch.save(val_tensor, os.path.join(output_dir, f"cmr_tensors_ts_all_val.pt"))
torch.save(test_tensor, os.path.join(output_dir, f"cmr_tensors_ts_all_test.pt"))
```
